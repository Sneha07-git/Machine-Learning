# Which one you should use? pyplot vs OO method
### When plotting something quickly it is okay to use pyplot method
### When plotting something more advanced use OO method

```python
heart_disease.head()
```

```python
over50 = heart_disease[heart_disease['age'] > 50]
over50.head()
```

```python
over50.plot(kind = 'scatter',x = 'age', y = 'chol', c = 'target');
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10,8))
over50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target',
           ax=ax);
ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over50['age'],
                    y=over50['chol'],
                    c=over50['target']);

# Customize the plot
ax.set(title='Age & Cholesterol Levels (Heart Diseases)',
      xlabel='Age',
      ylabel='Cholesterol');

# Add Legend
ax.legend(*scatter.legend_elements(),title='Target');

# Add Horizontal Line 
ax.axhline(over50['chol'].mean(),
          linestyle='--');
```

```python
# Subplot of chol, age & thalach
plt.style.use('seaborn-whitegrid')
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True); 


# Add data to ax0
scatter = ax0.scatter(x=over50['age'],
                     y=over50['chol'],
                     c=over50['target'],
                     cmap="winter");
# Customizing ax0
ax0.set(title='Age & Cholesterol Levels (Heart Diseases)',
      ylabel='Cholesterol');

# Setting x limit
ax0.set_xlim(50,80)

# Adding legends ax0
ax0.legend(*scatter.legend_elements(), title = 'Target');

# Adding horizontal line ax0
ax0.axhline(over50['chol'].mean(),
           linestyle='--',
           color='red');

# Add data to ax1
scatter = ax1.scatter(x=over50['age'],
                     y=over50['thalach'],
                     c=over50['target'],
                     cmap='winter');
# Customizing ax0
ax1.set(title='Age & Thalach Levels (Heart Diseases)',
      xlabel='Age',
      ylabel='Thalach');

# Setting x limit
ax1.set_ylim(60,200)

# Adding legends
ax1.legend(*scatter.legend_elements(), title = 'Target');

# Adding horizontal line 
ax1.axhline(over50['thalach'].mean(),
           linestyle='--',
           color='red');

# Adding title to entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

```
